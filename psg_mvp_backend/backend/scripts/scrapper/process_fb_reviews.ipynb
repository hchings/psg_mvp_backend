```python
import os
from os import listdir
from os.path import isfile, join
import json
import pandas as pd
from pathlib import Path

from nlp_engine.ner.procedure import extract_procedures, extract_procedures_bulk
```

```python
root = './puppeteer-login-for-facebook/fb_review_data'

onlyfiles = [f for f in listdir(root) if isfile(join(root, f)) and not f.startswith( 'error' )]
```

```python
dfs = []
print("Processing...", end = '')
for idx, file in enumerate(onlyfiles):
#     if idx != 41:
#         continue

    print("%s %s"%(idx,file), end = '')
    
    full_path = os.path.join(root, file)
    #print(full_path)
    
    with open(full_path) as f:
        data = json.load(f)
    #print(data)

    uuid = data['uuid']
    print('uuid %s'%(uuid))

    if not data["reviews"]:
        continue
        
    df = pd.DataFrame.from_dict(data["reviews"])
    df['uuid'] = uuid
    dfs.append(df)

df_all = pd.concat(dfs)
print(df_all.shape)
df_all.head()
df = extract_procedures_bulk(df_all)
df.to_csv('./fb_processed_reviews.csv', encoding='utf-8')
print('done~')


```
